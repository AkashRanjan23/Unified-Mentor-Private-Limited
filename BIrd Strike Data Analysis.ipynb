```python
# Dropping Some irrelevant Columns and Proceding :
new_data=bird_strike_data.drop(columns=['Record ID','Aircraft: Type','Airport: Name','Wildlife: Number struck','Remarks','Wildlife: Species'])
```

```python
new_data = new_data[new_data['Aircraft: Number of engines?'] != 'C']
```

```python
new_data.isna().sum()
```

```python
# Dropping Rows with Null Vlaues in the Column:
new_data.dropna(subset=['Altitude bin', 'FlightDate'], inplace=True)
```

```python
# Filling Missing values with Appropiate method:
new_data['Effect: Impact to flight'].fillna(method='ffill', inplace=True)

new_data['Aircraft: Number of engines?'] = pd.to_numeric(new_data['Aircraft: Number of engines?'], errors='coerce') # Converts the Column DataType from string to Intiger:
new_data['Aircraft: Number of engines?'].fillna(new_data['Aircraft: Number of engines?'].mean(), inplace=True)

mode_value = new_data[new_data['Origin State'] != 'N/A']['Origin State'].mode()[0]
new_data['Origin State'].fillna(mode_value, inplace=True)

mode_value = new_data[new_data['Conditions: Precipitation'] != 'None']['Conditions: Precipitation'].mode()[0]
new_data['Conditions: Precipitation'].fillna(mode_value, inplace=True)
```

```python
new_data.isna().sum()
```

```python
null_col=new_data.columns[new_data.isnull().any()].tolist()
null_col
```

```python
#Convert variable types and creating additional date variables
new_data['FlightDate'] = pd.to_datetime(new_data['FlightDate'])

# Extract the year
new_data['Year'] = new_data['FlightDate'].dt.year

# Extract the month (as a number)
new_data['Month'] = new_data['FlightDate'].dt.month

# Extract the month name
new_data['MonthName'] = new_data['FlightDate'].dt.month_name()

# Changing the Data Type to 'Int'
new_data['Aircraft: Number of engines?']=new_data['Aircraft: Number of engines?'].astype(int)
```

```python
# Convert 'Cost: Total $' to numeric, coercing errors to NaN
new_data['Cost: Total $'] = pd.to_numeric(new_data['Cost: Total $'], errors='coerce')

# Fill NaN values with a default value (e.g., 0) if necessary
new_data['Cost: Total $'].fillna(0, inplace=True)

# Convert the column to integers
new_data['Cost: Total $'] = new_data['Cost: Total $'].astype(int)
```

```python
def get_season(month):
    if month >= 3 and month <= 5:
        return 'Spring'
    elif month >= 6 and month <= 8:
        return 'Summer'
    elif month >= 9 and month <= 11:
        return 'Fall'
    else:
        return 'Winter'
    
new_data['Season']=new_data['Month'].apply(get_season)
```

# ***Analysis of the Impact of Weather Conditions on Bird Strikes and Resulting Aircraft Damage :***
